```python
import numpy as np
```

```python
data = open("inputs/12.input").read().strip()

ex_data = """
initial state: #..#.#..##......###...###

...## => #
..#.. => #
.#... => #
.#.#. => #
.#.## => #
.##.. => #
.#### => #
#.#.# => #
#.### => #
##.#. => #
##.## => #
###.. => #
###.# => #
####. => #
""".strip()
```

```python
lines = data.split("\n")
initial_state = lines[0][15:]
rules = {
    rule[:5]: rule[-1] for rule in lines[2:]
}
```

```python
initial_state
```

```python
plants = np.array(list(initial_state))
# create some space to the left and right, and save the offset to the '0' index
index_offs = 50
plants = np.pad(plants, index_offs, "constant", constant_values=".")
print("".join(plants))
for generation in range(20):
    next_gen = plants.copy()
    for i in range(2, len(plants) - 2):
        pattern = "".join(plants[i-2:i+2+1])
        next_gen[i] = rules[pattern]
        #next_gen[i] = "#" if pattern in rules else "."
    plants = next_gen
    print("".join(plants))
```

```python
pot_numbers = np.where(plants=="#")[0] - index_offs
pot_numbers.sum()
```

## Part 2

```python
plants = np.array(list(initial_state))
# create some space to the left and right, and save the offset to the '0' index
index_offs = 5
plants = np.pad(plants, index_offs, "constant", constant_values=".")
#print("".join(plants))
for generation in range(1000):
    if 990 <= generation <= 1000:
        print(generation, (plants=="#").sum())
        print("".join(plants))
    next_gen = plants.copy()
    for i in range(2, len(plants) - 2):
        pattern = "".join(plants[i-2:i+2+1])
        next_gen[i] = rules[pattern]
        #next_gen[i] = "#" if pattern in rules else "."
    plants = next_gen
    
    # dynamically grow the array (in both sides) if a plant is close to the edge
    if any(plants[:5] == "#") or any(plants[-5:] == "#"):
        index_offs += 5
        plants = np.pad(plants, 5, "constant", constant_values=".")
    
```

## After 1000 generations it has stabilized, and the plants just move one to the right every generation

```python
remaining_gens = 50000000000 - 1000
```

```python
offs = index_offs - remaining_gens
```

```python
pot_numbers = np.where(plants=="#")[0] - offs
pot_numbers.sum()
```
